In the chunk below I reset the images to be 224 from 460

```python
from fastai.vision.all import *
path = untar_data(URLs.PETS)
dls = ImageDataLoaders.from_name_re(path, get_image_files(path/'images'), pat='(.+)_\d+.jpg', item_tfms=Resize(224), batch_tfms=aug_transforms(size=224, min_scale=0.75))
learn = vision_learner(dls, models.resnet50, metrics=accuracy)
learn.fine_tune(1)
learn.path = Path('.')
learn.export()
```

```python
learn = load_learner('export.pkl')
```

```python
labels = learn.dls.vocab

def predict(img):
    img = PILImage.create(img)
    pred,pred_idx,probs = learn.predict(img)
    return {labels[i]: float(probs[i]) for i in range(len(labels))}
```

```python
demo.close()
```

```python
import gradio as gr
demo = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil"),          # or type="numpy" if your predict expects arrays
    outputs=gr.Label(num_top_classes=3),
)

demo.launch(share=True)
```
